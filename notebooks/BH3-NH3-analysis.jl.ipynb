# BH3-NH3 Electronic Density Analysis

Analyzing the CSV output from a PES scan.

## Dataset Overview: Gradient Bundle Analysis

This dataset contains point-wise function values along 1D "gradient paths" embedded in 3D space. These paths trace the gradient of the electron charge density within the chemical system, starting from "seeds" arranged regularly on a sphere around specific atomic coordinates. 

The paths serve to approximate the volumetric properties of the surrounding region, known as a **differential gradient bundle**.

### Key Columns:
* **`Atom_Label` & `GB_Index`**: Identify the specific atom and the path (gradient bundle index). Each combination of these two values represents a unique path consisting of multiple points.
* **`arc_length`**: The distance along the path, serving as the independent variable (x-axis) for analysis.
* **`closest saddle distance`**: An example of a dependent variable that can be plotted against `arc_length` for specific paths.
